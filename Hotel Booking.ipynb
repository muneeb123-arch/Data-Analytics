```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Hotel_booking=pd.read_excel ('hotel_booking.xlsx')
```

```python
avg_lead_time = Hotel_booking.groupby('hotel')['lead_time'].mean()

# Step 3: Create a pie chart
plt.figure(figsize=(8, 8))  # Set the figure size
plt.pie(avg_lead_time, labels=avg_lead_time.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Average Lead Time by Hotel')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Step 4: Display the pie chart
plt.show()
```

```python
# Step 2: Calculate the average lead time for each hotel by month
avg_lead_time_by_month = Hotel_booking.groupby(['arrival_date_month', 'hotel'])['lead_time'].mean().unstack()

# Step 3: Create the clustered bar chart
ax = avg_lead_time_by_month.plot(kind='bar', figsize=(12, 7), width=0.8)

# Step 4: Add labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Step 5: Customize the plot
plt.title('Average Lead Time by Arrival Month and Hotel')
plt.xlabel('Arrival Month')
plt.ylabel('Average Lead Time (days)')
plt.legend(title='Hotel', loc='upper right')
plt.xticks(rotation=45)  # Rotate month labels for better readability
plt.tight_layout()  # Adjust layout to ensure everything fits

# Step 6: Display the plot
plt.show()
```

```python
# Assuming 'hotel_booking' is your DataFrame and 'reservation_status' is the column that contains statuses.
# The status "Canceled" indicates canceled bookings.

# Calculate the total number of bookings
total_bookings = Hotel_booking['reservation_status'].count()

# Calculate the number of canceled bookings
canceled_bookings = Hotel_booking[Hotel_booking['reservation_status'] == 'Canceled']['reservation_status'].count()

# Calculate the cancellation rate
cancellation_rate = 100*(canceled_bookings / total_bookings)
```

```python
hotel_data = Hotel_booking.groupby('hotel').agg(
    total_bookings=('reservation_status', 'count'),
    canceled_bookings=('reservation_status', lambda x: (x == 'Canceled').sum())
)

# Calculate the cancellation rate for each hotel
hotel_data['cancellation_rate'] = 100*hotel_data['canceled_bookings'] / hotel_data['total_bookings']

# Plotting the cancellation rate by hotel
plt.figure(figsize=(10, 6))
hotel_data['cancellation_rate'].plot(kind='bar', color='skyblue')

# Add labels and title
plt.title('Cancellation Rate by Hotel', fontsize=16)
plt.xlabel('Hotel', fontsize=12)
plt.ylabel('Cancellation Rate (%)', fontsize=12)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a new column for cancellation rate: 1 for "Canceled", 0 for others
Hotel_booking['cancellation_rate'] = Hotel_booking['reservation_status'].apply(lambda x: 1 if x == 'Canceled' else 0)

# Group by hotel and month to calculate average cancellation rate and lead time
hotel_month_data = Hotel_booking.groupby(['hotel', 'arrival_date_month']).agg(
    average_lead_time=('lead_time', 'mean'),
    cancellation_rate=('cancellation_rate', 'mean')
).reset_index()

# Convert the cancellation rate to percentage
hotel_month_data['cancellation_rate'] = hotel_month_data['cancellation_rate'] * 100

# Set the plot size
plt.figure(figsize=(10, 6))

# Create scatter plot using seaborn with 'cancellation_rate' on x-axis and 'average_lead_time' on y-axis
sns.scatterplot(data=hotel_month_data, 
                x='cancellation_rate', 
                y='average_lead_time', 
                hue='arrival_date_month', 
                palette='viridis', 
                s=100)  # Size of scatter points

# Add a regression line (optional)
sns.regplot(data=hotel_month_data, 
            x='cancellation_rate', 
            y='average_lead_time', 
            scatter=False, 
            color='black', 
            line_kws={"linestyle":"--"})

# Add labels and title
plt.title('Cancellation Rate (Percentage) and Average Lead Time by Arrival Month', fontsize=16)
plt.xlabel('Cancellation Rate (%)', fontsize=12)
plt.ylabel('Average Lead Time (Days)', fontsize=12)

# Rotate x-axis labels if necessary (for better readability)
plt.xticks(rotation=45)

# Display the legend
plt.legend(title='Arrival Month', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
